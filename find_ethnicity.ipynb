# Exctract meaning of ethnicity code

```python
import pandas as pd
from webscraping.converter import EntityConverter
```

```python
actor_df = pd.read_csv('data/actor_data.csv', index_col=0)
test_sample = actor_df[:100]
test_sample
```

```python
entity_converter = EntityConverter("https://query.wikidata.org/sparql") # use the Wikidata SPARQL endpoint

# example use of the converter
entity_converter.get_ethnicity("/m/02p656s")
```

```python
test_sample.loc[:, 'Actor ethnicity'] = test_sample['Actor ethnicity'].apply(entity_converter.get_ethnicity)
test_sample
```
