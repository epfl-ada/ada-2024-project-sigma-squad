```python
import pandas as pd
```

```python
actor_df = pd.read_csv('actor_data_for_regression.csv', index_col=0)
actor_df
```

```python
missing_birth_year = actor_df.loc[actor_df['Birth Year'].isna(),'Birth Year']

print(f"Actors with missing birth years: {len(missing_birth_year)}")
for actor in missing_birth_year.index:
    print(actor)
```

```python
missing_birthdays = {
    'Jeremy Irvine': (1990, 'June'),
    'Rolf Saxon': (1955, 'July'),
}

for actor, (year, month) in missing_birthdays.items():
    actor_df.loc[actor, 'Birth Year'] = year
    actor_df.loc[actor, 'Birth Month'] = month
```
