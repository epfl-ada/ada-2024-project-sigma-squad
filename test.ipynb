```python
import pandas as pd
import plotly.io as pio
import plotly.express as px
```

```python
states_df = pd.read_csv('state_points_for_plot.csv')
actor_df = pd.read_csv('plot_ready_actor_df.csv')
```

```python
fig = px.scatter(
    actor_df,
    x="Usable Uni Rank",
    y="Success Score",
    color='University',
    title="Usable Uni Rank vs. Success Score",
    hover_data={"index": actor_df.index}  # Include index in hover data
)

fig.show()
fig.write_html("rank_fixed_vs_score.html")
```

```python
# Create a choropleth map
fig = px.choropleth(
    states_df,
    locations="Birth State",
    locationmode="USA-states",  # Use USA states mode
    color="Success Score",
    scope="usa",  # Restrict map to USA
    title="Cumulative Random Draw Success Score by Birth State",
    color_continuous_scale="Reds",  # You can choose other color scales
    labels={"Success Score": "Cumul. Success Score"}
)

# Show the map
fig.show()
fig.write_html("state_vs_score.html")
```
