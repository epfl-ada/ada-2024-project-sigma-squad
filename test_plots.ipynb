```python

import requests
from bs4 import BeautifulSoup
import re
import time

def scrape_actor_education_with_bs(actor_df):
    """
    Scrapes Wikipedia to determine if actors majored in acting, drama, or fine arts
    and updates the 'University' column based on proximity to keywords.

    Args:
        actor_df (pd.DataFrame): DataFrame with actor names as the index and a 'University' column.

    Returns:
        pd.DataFrame: Updated DataFrame with modified 'University' values.
    """
    # Base Wikipedia URL
    wiki_base_url = "https://en.wikipedia.org/wiki/"
    
    # Helper function to check word proximity
    def is_keyword_near(text, main_keywords, proximity_keywords, max_distance=5):
        words = re.findall(r'\w+', text.lower())
        for i, word in enumerate(words):
            if word in main_keywords:
                for j in range(max(0, i - max_distance), min(len(words), i + max_distance + 1)):
                    if words[j] in proximity_keywords:
                        return True
        return False

    # Keywords to search for
    majors = ['major', 'majored', 'classes']
    fields = {
        'Specialised Acting School': ['acting'],
        'Specialised Drama School': ['drama'],
        'Specialised Arts School': ['fine', 'arts']
    }
    
    # Iterate through the DataFrame
    for actor in actor_df.index:
        # Create the Wikipedia page URL
        actor_name = actor.replace(" ", "_")
        actor_url = wiki_base_url + actor_name

        try:
            # Request the page content
            response = requests.get(actor_url, headers={"User-Agent": "ActorEducationScraper/1.0 (contact: aiden.tschammer.ato@gmail.com)"})
            response.raise_for_status()
            
            # Parse the HTML content with BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Find the "Early life" section
            early_life_header = soup.find('h2', id=re.compile('Early_life', re.IGNORECASE))
            if not early_life_header:
                continue  # Skip if no Early Life section
            print('found early_life')
            # Extract the text of the "Early life" section
            early_life_text = ""
            for sibling in early_life_header.parent.find_next_siblings():
                if sibling.name == "h2":  # Stop at the next main section
                    break
                early_life_text += sibling.get_text(separator=" ")

            # Check for proximity of keywords
            for school, field_keywords in fields.items():
                if is_keyword_near(early_life_text, field_keywords, majors, max_distance=5):
                    actor_df.loc[actor, 'University'] = school
                    break

            # Respect Wikipedia's servers with a delay
            time.sleep(2)

        except requests.RequestException as e:
            print(f"Error fetching data for {actor}: {e}")
    
    return actor_df



 
```

```python
def groupby_states(actor_df):

    states = [
        'Alabama',
        'Alaska',
        'Arizona',
        'Arkansas',
        'California',
        'Colorado',
        'Connecticut',
        'Delaware',
        'Florida',
        'Georgia',
        'Hawaii',
        'Idaho',
        'Illinois',
        'Indiana',
        'Iowa',
        'Kansas',
        'Kentucky',
        'Louisiana',
        'Maine',
        'Maryland',
        'Massachusetts',
        'Michigan',
        'Minnesota',
        'Mississippi',
        'Missouri',
        'Montana',
        'Nebraska',
        'Nevada',
        'New Hampshire',
        'New Jersey',
        'New Mexico',
        'New York',
        'North Carolina',
        'North Dakota',
        'Ohio',
        'Oklahoma',
        'Oregon',
        'Pennsylvania',
        'Rhode Island',
        'South Carolina',
        'South Dakota',
        'Tennessee',
        'Texas',
        'Utah',
        'Vermont',
        'Virginia',
        'Washington',
        'West Virginia',
        'Wisconsin',
        'Wyoming'
        ]
    
    regions = ['USA', 'United Kingdom', 'Europe', 'nan']

    states_df = pd.DataFrame(index=actor_df.index, columns=['Birth State'])

    states_df['Birth State'] = actor_df['Birth City'].astype(str)

    for state in states:
        states_df.loc[states_df['Birth State'].str.contains(state, case=False), 'Birth State'] = state

    states_df['Birth State'] = states_df['Birth State'].apply(lambda x: x if x in states else None)

    return states_df
```

```python
states_df = groupby_states(actor_df)
states_df['Success Score'] = actor_df['Success Score'].copy()
```

```python
states_df.dropna(inplace=True)
states_df
```

```python
# Define grade ranges and corresponding weights
grade_ranges = [(10, 8), (8, 6), (6, 4), (4, 2), (2, 0)]
weights = [1, 0.75, 0.5, 0.25, 0]

# Initialize state points dictionary
state_points = {state: 0 for state in states_df['Birth State'].unique()}

# Number of iterations
iterations = 1000

# Perform the selection and scoring
for _ in range(iterations):
    selected_states = []
    for grade_range, weight in zip(grade_ranges, weights):
        # Filter actors within the current grade range
        filtered_df = states_df[(states_df['Success Score'] <= grade_range[0]) & (states_df['Success Score'] > grade_range[1])]
        
        # Randomly select one actor (if available)
        if not filtered_df.empty:
            selected_actor = filtered_df.sample(1).iloc[0]
            selected_states.append((selected_actor['Birth State'], weight))
    
    # Assign points to states
    for state, weight in selected_states:
        state_points[state] += weight

# Convert state_points to a DataFrame for visualization
state_points_df = pd.DataFrame(list(state_points.items()), columns=['Birth State', 'Success Score'])
```
